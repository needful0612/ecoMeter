### c. univariate analysis(single feature)

```python
import matplotlib.pyplot as plt
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

print("Numeric columns:", numeric_cols)
print("Categorical columns:", categorical_cols)
```

```python
df['date'] = pd.to_datetime(df['date'])
df['hour'] = df['date'].dt.hour
numeric_cols = df.select_dtypes(include=['int64', 'float64', 'int32']).columns
len(numeric_cols)
```

```python
n_cols = 4
n_rows = (len(numeric_cols) + n_cols - 1) // n_cols

plt.figure(figsize=(n_cols*4, n_rows*3)) 

for i, col in enumerate(numeric_cols):
    plt.subplot(n_rows, n_cols, i+1) 
    df[col].hist(bins=40)
    plt.title(col, fontsize=10)
    plt.xlabel('')
    plt.ylabel('')

plt.tight_layout()
plt.show()
```

### d. Bivariate analysis

```python
n_cols = 4
n_rows = (len(numeric_cols) + n_cols - 1) // n_cols

plt.figure(figsize=(n_cols * 5, n_rows * 4))

numeric_cols_copy = numeric_cols.drop('Appliances')

for i, col in enumerate(numeric_cols_copy):
    plt.subplot(n_rows, n_cols, i + 1)
    plt.scatter(df[col], df['Appliances'], alpha=0.3)
    plt.title(f"{col} vs Appliances", fontsize=9)
    plt.xlabel(col, fontsize=8)
    plt.ylabel("Appliances", fontsize=8)

plt.tight_layout()
plt.show()
```

### e. Multivariate analysis 

```python
import seaborn as sns
```

```python
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(14, 12))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python
# w/o appliance
corr_matrix = df[numeric_cols_copy].corr()

plt.figure(figsize=(14, 12))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```
